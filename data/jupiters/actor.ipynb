```python
def find_dipli(df, full_name):
    """
    Args:
        df: The DataFrame containing the data.
        full_name: The name of the column you want to check for duplicates.
        
    Returns:
        A DataFrame with the duplicated rows based on the 'full_name' column.
    """
    return df[df[full_name].duplicated()]
```

```python
dupli_ = find_dipli(df, 'full_name')

print(dupli_)
```
